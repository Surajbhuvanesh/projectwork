```python
##Import numpy and pandas 
##Import visualization libraries and set %matplotlib inline.
import numpy as np
import pandas as pd
import matplotlib as mlp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##Read in the csv file as a dataframe called df
call=pd.read_csv('911.csv')
call
```

```python
##Check the info() of the df
call.info()
```

```python
##Check the head of df
call.head()
```

```python
##What are the top 5 zipcodes for 911 calls?
call.fillna(0)
call['zip'].value_counts().head()
```

```python
##What are the top 5 townships (twp) for 911 calls?
call['twp'].value_counts()[:5]
```

```python
##Take a look at the 'title' column, how many unique title codes are there?
a=call['title']
b=a.unique()
b.size
```

```python
##In the titles column there are "Reasons/Departments" specified before the title code. These are EMS, Fire, and Traffic. Use .apply() with a custom lambda expression to create a new column called "Reason" that contains this string value.For example, if the title column value is EMS: BACK PAINS/INJURY , the Reason column value would be EMS.
call["Reason"]=call["title"].apply(lambda x : x.split(":")[0])
call
```

```python
##What is the most common Reason for a 911 call based off of this new column?
call['Reason'].value_counts()[:5]
```

```python
##Now use seaborn to create a countplot of 911 calls by Reason.
import seaborn as sns
sns.countplot(x=call['Reason'],data=call)
```

```python
##Now let us begin to focus on time information. What is the data type of the objects in the timeStamp column?
type(call['timeStamp'][0])
```

```python
##You should have seen that these timestamps are still strings. Use pd.to_datetime to convert the column from strings to DateTime objects
call['timeStamp']=pd.to_datetime(call['timeStamp'])
type(call['timeStamp'][0])
```

```python
##You can use Jupyter's tab method to explore the various attributes you can call. Now that the timestamp column are actually DateTime objects, use .apply() to create 3 new columns called Hour, Month, and Day of Week. You will create these columns based off of the timeStamp column, reference the solutions if you get stuck on this step
call['Hour']=call['timeStamp'].apply(lambda x : x.hour)
call["Month"]=call["timeStamp"].apply(lambda x: x.month)
call["Day of Week"]=call["timeStamp"].apply(lambda x: x.dayofweek)
call.head()
```

```python
##Notice how the Day of Week is an integer 0-6. Use the .map() with this dictionary to map the actual string names to the day of the week
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
call["Day of Week"]= call["Day of Week"].map(dmap)
call.head()
```

```python
##Now use seaborn to create a countplot of the Day of Week column with the hue based off of the Reason column.
sns.countplot(data=call,x='Day of Week',hue=call['Reason'])
plt.legend(loc='upper left',bbox_to_anchor=(1,0.5))
```

```python
##Now do the same for Month:
sns.countplot(data=call,x='Month',hue=call['Reason'])
plt.legend(loc='upper left',bbox_to_anchor=(1,0.5))
```

```python
##Now create a gropuby object called byMonth, where you group the DataFrame by the month column and use the count() method for aggregation. Use the head() method on this returned DataFrame.
byMonth=call.groupby("Month").count()
byMonth.head()
```

```python
byMonth['addr'].plot()
```

```python
##Now see if you can use seaborn's lmplot() to create a linear fit on the number of calls per month. Keep in mind you may need to reset the index to a column.
from seaborn import lmplot
sns.lmplot(data=byMonth.reset_index(),x='Month',y='twp')
```

```python
##Create a new column called 'Date' that contains the date from the timeStamp column. You'll need to use apply along with the .date() method.
call['Date']=call['timeStamp'].apply(lambda x : x.date())
call.head(3)
```

```python
##Now groupby this Date column with the count() aggregate and create a plot of counts of 911 calls.
d=call.groupby('Date').count()
d['lat'].plot()
```

```python
##Now recreate this plot but create 3 separate plots with each plot representing a Reason for the 911 call
call[call['Reason']=='Traffic'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
```

```python
call[call['Reason']=='Fire'].groupby('Date').count()['twp'].plot()
plt.title('Fire')
```

```python
call[call['Reason']=='EMS'].groupby('Date').count()['twp'].plot()
plt.title('EMS')
```

```python
Hd=call.groupby(["Day of Week","Hour"]).count()["Reason"].unstack(level=-1)
Hd.head()
```

```python
##Now create a HeatMap using this new DataFrame
sns.heatmap(Hd,cmap='rainbow')
```

```python
##Now create a clustermap using this DataFrame
sns.clustermap(Hd,cmap='rainbow')
```

```python
##Now repeat these same plots and operations, for a DataFrame that shows the Month as the column.
m=call.groupby(['Day of Week','Month']).count()['Reason'].unstack(level=-1)
m.head()
```

```python
sns.heatmap(m,cmap='rocket')
```

```python
sns.clustermap(m,cmap='spring')
```
